# 3. Изменяем дата-фрейм

### 3.1 Удаляем данные

```python
# .isnull() - поиск значений NaN
df.isnull()
```

```python
# суммируем все найденные значения NaN
df.isnull().sum()
```

```python
# .dropna удаляем пропущенные значения в серии
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

### 3.2 Удаляем столбцы

```python
# удаляем столбец "Area code"
df.drop('Area code', axis=1)
```

### 3.3 Создание столбца с суммой других столбцов

```python
# запись в столбец суммы других столбцов
df['New Column'] = df['Total night minutes'] + df['Total intl minutes']
```

```python
df.head()
```

### 3.4 Заполнение столбца данными

```python
# заполнение столбца 100
df['New Column'] = 100
```

```python
df.head()
```

### 3.5 Изменение содержимого ячейки

```python
# меняем значение конкретной ячейки
df.iloc[0,-1] = 10
```

```python
df.head()
```

### 3.6 Создаём столбец на основе данных другого столбца

```python
# применяем лямбда-функцию
df['Churn Binary'] = df['Churn'].apply(lambda x: 1 if x==True else 0)
```

```python
df[df['Churn']==True].head()
```
